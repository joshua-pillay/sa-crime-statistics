# Stage 7 - SQL Summary View Exporting

#### Queries the database populated with the summary view and exports it to file by:

- Loading and connecting to the populated `sa_crime` database from `data/sql/sa_crime.db`,
- Querying the summary view with queries executed against `sa_crime.db` to a dataframe `summary_view_df`, and
- Exporting the summary dataframe `summary_view_df` CSV to `sa_crime_summary.csv` for the dashboarding stage in the project pipeline.

*Note - Execute this notebook after running `schema.sql` and `queries.sql` scripts (in `sql/`) against `sa_crime_db`.*
#### Data source: [Crime Statistics of South Africa (2011-2023)](https://www.kaggle.com/datasets/harutyunagababyan/crime-stats-of-south-africa-2011-2023)

---------------

Imports necessary libraries: `pandas` for data manipulation, `sqlite3` for database operations, and `os` for directory navigation:

```python
import pandas as pd
import sqlite3
import os 
```

Navigates from notebook's location to project root, i.e. since notebooks live in `notebooks/`, goes up one level to reach project root:

```python
dir_nb = os.getcwd()
project_root = os.path.dirname(dir_nb)    
os.chdir(project_root)                   

print(os.getcwd())                     
```

Imports populated database from file and extracts summary view to CSV:

```python
conn = sqlite3.connect('data/sql/sa_crime.db')       # connects to populated database

summary_view_df =  pd.read_sql_query("SELECT * FROM sa_crime_summary", conn)     # queries summary view
 
summary_view_df.to_csv('data/processed/sa_crime_summary.csv', index = False)     # exports summary summary to csv

conn.close()                                         # closes connection
```

The output of this notebook is as follows: `sa_crime_summary.csv` for the dashboarding stage in the project pipeline.
